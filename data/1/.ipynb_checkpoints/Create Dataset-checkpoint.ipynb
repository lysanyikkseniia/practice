```python
import pandas as pd
```

```python
def read_df(year):
    csv_name = 'dst'+year+'_hv.txt'

    df = pd.read_csv(csv_name, header=None)

    df['Дата'] = year + '-' + df[0].astype(str).str[5:7] + '-' + df[0].astype(str).str[8:10]

    df['Базисное значение'] = df[0].astype(str).str[17:20]

    df['Базисное значение'] = pd.to_numeric(df['Базисное значение'])

    df['Среднесуточное значение'] = df[0].astype(str).str[-3:]

    df['Среднесуточное значение'] = pd.to_numeric(df['Среднесуточное значение'])

    l = 21
    r = 24

    for x in range(25):
        df[str(x) + ':00'] = df[0].astype(str).str[l:r]
        df[str(x) + ':00'] = pd.to_numeric(df[str(x) + ':00'])
        l += 4
        r += 4

    df.drop([0], axis = 1, inplace = True)
    
    return df
```

```python
def make_year_df(df):

    df1 = pd.DataFrame(columns=['Дата','Базисное значение','Среднесуточное значение', 'Текущее значение'])

    day_number = 0

    count = 0

    while day_number < len(df.index):

        for hour in range(25):
            df1.loc[count] = pd.Series({'Дата':df['Дата'][day_number] + ' ' + str(hour) + ':00', 'Базисное значение':df['Базисное значение'][day_number], 'Среднесуточное значение':df['Среднесуточное значение'][day_number], 'Текущее значение':df[str(hour) + ':00'][day_number]})

            count += 1

        day_number += 1
    
    return df1
```

```python
df = pd.DataFrame()
for i in range(1957, 2015):
    df_year = make_year_df(read_df(str(i)))
    df = pd.concat([df, df_year], ignore_index=True)
```

```python
df.to_csv('dst_data.csv', index=False)
```
